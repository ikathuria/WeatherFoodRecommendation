## Merging gtrends data

```python
import os
import pandas as pd
```

```python
maindir = "gtrends_data/"

res = pd.read_csv("gtrends_data/Aloo gobi.csv", skiprows=1)
for i in os.listdir(maindir)[1:]:
    filename = maindir + i
    res = pd.concat([res, pd.read_csv(filename, skiprows=1,
                    usecols=lambda x: x != 'Week')], axis=1)
res = res.fillna(0)
res
```

```python
res.index = pd.to_datetime(res.Week)
res.drop('Week', axis=1, inplace=True)
res
```

```python
res.to_excel("delhi_food_trends_merged.xlsx")
```

## Converting Delhi temperatures from daily to weekly averages

```python
df = pd.read_excel("delhi_temperature.xlsx").fillna(0)
df.rename({"date": "Week"}, axis=1, inplace=True)
df["Week"] = pd.to_datetime(df["Week"])
df
```

```python
def avrg(values):
  return sum(values) / len(values)
```

```python
result = df.groupby(pd.Grouper(key='Week', freq='W')).agg({
    'tavg': avrg,
    'tmin': avrg,
    'tmax': avrg,
    'prcp': avrg
})
result = result[:-1]
result
```

```python
result.to_excel("delhi_temperate_weekly.xlsx")
```

## Merging both files to get final data

```python
result = pd.read_excel("delhi_temperate_weekly.xlsx", index_col=0)
res = pd.read_excel("delhi_food_trends_merged.xlsx", index_col=0)

pd.concat([result, res], axis=1).to_excel("delhi_temperature_trends_merged.xlsx")
```
